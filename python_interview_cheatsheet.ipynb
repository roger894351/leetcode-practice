## 1. Lists & Arrays

```python
# ── Creating & basics ──────────────────────────────────────────────────────
nums = [3, 1, 4, 1, 5, 9, 2, 6]

print(nums[0])        # first element  → 3
print(nums[-1])       # last element   → 6
print(nums[1:4])      # slice [1,4)    → [1, 4, 1]
print(nums[::-1])     # reversed       → [6, 2, 9, 5, 1, 4, 1, 3]
print(len(nums))      # length         → 8
```

```python
# ── Mutating ───────────────────────────────────────────────────────────────
nums = [1, 2, 3]
nums.append(4)          # [1, 2, 3, 4]    O(1)
nums.insert(1, 99)      # [1, 99, 2, 3, 4] O(n)  ← avoid in hot loops
nums.pop()              # removes last     O(1)
nums.pop(0)             # removes index 0  O(n)  ← use deque if frequent
nums.remove(99)         # removes by value O(n)
print(nums)
```

```python
# ── Iteration patterns ─────────────────────────────────────────────────────
nums = [10, 20, 30]

for i in range(len(nums)):          # index only
    print(i, end=' ')               # 0 1 2
print()

for i, val in enumerate(nums):      # index + value  ← prefer this
    print(i, val)

for a, b in zip([1,2,3], [4,5,6]): # pair two lists
    print(a, b)
```

```python
# ── List comprehensions ────────────────────────────────────────────────────
squares   = [x**2 for x in range(6)]           # [0,1,4,9,16,25]
evens     = [x for x in range(10) if x % 2==0] # [0,2,4,6,8]
matrix    = [[0]*3 for _ in range(3)]           # 3×3 zero matrix  ← safe!
# WRONG way: [[0]*3]*3  — all rows share the same reference!

print(squares)
print(evens)
print(matrix)
```

```python
# ── Useful list methods ────────────────────────────────────────────────────
nums = [3, 1, 4, 1, 5]
print(min(nums), max(nums), sum(nums))
print(nums.count(1))         # count occurrences → 2
print(nums.index(4))         # first index of value → 2
nums.sort()                  # in-place sort
print(nums)
print(sorted([3,1,2]))       # returns new sorted list
```

## 2. Strings

```python
# ── String basics ──────────────────────────────────────────────────────────
s = "Hello, World!"

print(s.lower())             # hello, world!
print(s.upper())             # HELLO, WORLD!
print(s.strip())             # remove whitespace from both ends
print(s.replace('World','Python'))  # Hello, Python!
print(s.split(', '))         # ['Hello', 'World!']
print(s.startswith('Hello')) # True
print(s.endswith('!'))       # True
print(len(s))                # 13
print('llo' in s)            # True  ← O(n) search
```

## 3. Dictionary / HashMap

```python
# ── Creating & accessing ───────────────────────────────────────────────────
d = {'a': 1, 'b': 2}
print(d['a'])                # 1  (KeyError if missing)
print(d.get('c', 0))        # 0  (safe default)  ← prefer this
print('a' in d)             # True
d['c'] = 3                  # insert / update
del d['a']                  # delete key
print(d)
```

```python
# ── Frequency count — most common interview pattern ───────────────────────
nums = [1, 2, 2, 3, 3, 3]

# Manual way
count = {}
for n in nums:
    count[n] = count.get(n, 0) + 1
print(count)                 # {1:1, 2:2, 3:3}

# Pythonic way
from collections import Counter
count = Counter(nums)
print(count)                 # Counter({3:3, 2:2, 1:1})
print(count.most_common(2))  # [(3,3),(2,2)]  ← top k elements
```

```python
# ── defaultdict — avoids KeyError ─────────────────────────────────────────
from collections import defaultdict

graph = defaultdict(list)    # default value is []
graph['a'].append('b')       # no need to check if 'a' exists
graph['a'].append('c')
print(dict(graph))           # {'a': ['b', 'c']}

freq = defaultdict(int)      # default value is 0
for ch in 'abracadabra':
    freq[ch] += 1
print(dict(freq))
```

```python
# ── Iterating a dict ───────────────────────────────────────────────────────
d = {'x': 10, 'y': 20, 'z': 30}

for k in d:                  # keys
    print(k, end=' ')
print()

for v in d.values():         # values
    print(v, end=' ')
print()

for k, v in d.items():       # both  ← most useful
    print(f'{k}={v}', end=' ')
```

## 6. Two Pointers

```python
# ── Pattern: opposite ends → meet in middle ────────────────────────────────
# Example: Two Sum II (sorted array)

def two_sum_sorted(nums, target):
    left, right = 0, len(nums) - 1
    while left < right:
        s = nums[left] + nums[right]
        if s == target:
            return [left, right]
        elif s < target:
            left += 1
        else:
            right -= 1
    return []

print(two_sum_sorted([2,7,11,15], 9))   # [0,1]
print(two_sum_sorted([2,3,4], 6))       # [0,2]
```

```python
# ── Pattern: fast / slow pointer ───────────────────────────────────────────
# Example: check if a number is happy (cycle detection)

def is_palindrome(s):
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

print(is_palindrome('racecar'))  # True
print(is_palindrome('hello'))    # False
```

## 7. Sliding Window

```python
# ── Fixed window: max sum of k consecutive elements ────────────────────────
def max_sum_fixed(nums, k):
    window_sum = sum(nums[:k])
    best = window_sum
    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i - k]   # slide: add right, drop left
        best = max(best, window_sum)
    return best

print(max_sum_fixed([2,1,5,1,3,2], 3))  # 9  (5+1+3)
```

```python
# ── Variable window: longest subarray with sum ≤ target ───────────────────
def longest_subarray(nums, target):
    left = 0
    window_sum = 0
    best = 0
    for right in range(len(nums)):
        window_sum += nums[right]            # expand right
        while window_sum > target:           # shrink left until valid
            window_sum -= nums[left]
            left += 1
        best = max(best, right - left + 1)   # window size
    return best

print(longest_subarray([4,2,2,7,8,1,2,8,1,0], 8))  # 3
```

## 8. Binary Search

```python
# ── Classic binary search ──────────────────────────────────────────────────
def binary_search(nums, target):
    left, right = 0, len(nums) - 1
    while left <= right:               # NOTE: <=  not <
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1                          # not found

print(binary_search([1,3,5,7,9], 7))  # 3
print(binary_search([1,3,5,7,9], 4))  # -1
```

```python
# ── bisect — built-in binary search ───────────────────────────────────────
import bisect

nums = [1, 3, 5, 7, 9]
print(bisect.bisect_left(nums, 5))    # 2  (leftmost insert position)
print(bisect.bisect_right(nums, 5))   # 3  (rightmost insert position)
bisect.insort(nums, 4)                # insert in sorted order
print(nums)                           # [1,3,4,5,7,9]
```

```python
# ── Binary search on answer space ─────────────────────────────────────────
# Example: find square root (integer) without sqrt()
def my_sqrt(n):
    left, right = 0, n
    ans = 0
    while left <= right:
        mid = (left + right) // 2
        if mid * mid <= n:
            ans = mid            # valid, try larger
            left = mid + 1
        else:
            right = mid - 1      # too big
    return ans

print(my_sqrt(8))   # 2
print(my_sqrt(9))   # 3
```

## 9. Stack & Queue

```python
# ── Stack — LIFO (use a list) ──────────────────────────────────────────────
stack = []
stack.append(1)     # push
stack.append(2)
stack.append(3)
print(stack[-1])    # peek top → 3
stack.pop()         # pop  → 3
print(stack)
```

```python
# ── Classic stack problem: valid parentheses ───────────────────────────────
def is_valid(s):
    stack = []
    mapping = {')':'(', '}':'{', ']':'['}
    for ch in s:
        if ch in mapping:                        # closing bracket
            if not stack or stack[-1] != mapping[ch]:
                return False
            stack.pop()
        else:                                    # opening bracket
            stack.append(ch)
    return len(stack) == 0

print(is_valid('()[]{}'))   # True
print(is_valid('([)]'))     # False
```

```python
# ── Queue — FIFO (use deque for O(1) popleft) ─────────────────────────────
from collections import deque

q = deque()
q.append(1)         # enqueue right
q.append(2)
q.appendleft(0)     # enqueue left
print(q)            # deque([0,1,2])
print(q.popleft())  # dequeue left → 0  O(1)
print(q.pop())      # dequeue right → 2  O(1)
print(q)
```

```python
# ── Monotonic stack — next greater element ─────────────────────────────────
def next_greater(nums):
    result = [-1] * len(nums)
    stack = []           # stores indices, maintains decreasing order
    for i, val in enumerate(nums):
        while stack and nums[stack[-1]] < val:
            idx = stack.pop()
            result[idx] = val
        stack.append(i)
    return result

print(next_greater([2,1,2,4,3]))  # [4,2,4,-1,-1]
```

## 10. Linked List

```python
# ── ListNode definition (used in LeetCode) ────────────────────────────────
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def make_list(vals):
    dummy = ListNode(0)
    cur = dummy
    for v in vals:
        cur.next = ListNode(v)
        cur = cur.next
    return dummy.next

def print_list(head):
    vals = []
    while head:
        vals.append(str(head.val))
        head = head.next
    print(' -> '.join(vals))

head = make_list([1,2,3,4,5])
print_list(head)
```

```python
# ── Reverse a linked list ──────────────────────────────────────────────────
def reverse_list(head):
    prev = None
    cur = head
    while cur:
        nxt = cur.next   # save next
        cur.next = prev  # reverse pointer
        prev = cur       # advance prev
        cur = nxt        # advance cur
    return prev

head = make_list([1,2,3,4,5])
print_list(reverse_list(head))  # 5->4->3->2->1
```

```python
# ── Dummy node pattern — simplifies edge cases ─────────────────────────────
# Example: delete node with given value
def delete_val(head, val):
    dummy = ListNode(0, head)   # dummy before head
    cur = dummy
    while cur.next:
        if cur.next.val == val:
            cur.next = cur.next.next  # skip the node
        else:
            cur = cur.next
    return dummy.next

head = make_list([1,2,3,2,4])
print_list(delete_val(head, 2))  # 1->3->4
```

## 11. Tree & DFS / BFS

```python
# ── TreeNode definition (LeetCode standard) ───────────────────────────────
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Build sample tree:    1
#                      / \
#                     2   3
#                    / \
#                   4   5
root = TreeNode(1,
    TreeNode(2, TreeNode(4), TreeNode(5)),
    TreeNode(3))
print("Tree built")
```

```python
# ── DFS traversals (recursive) ────────────────────────────────────────────
def inorder(node):    # Left → Root → Right  (BST gives sorted order)
    if not node: return []
    return inorder(node.left) + [node.val] + inorder(node.right)

def preorder(node):   # Root → Left → Right
    if not node: return []
    return [node.val] + preorder(node.left) + preorder(node.right)

def postorder(node):  # Left → Right → Root
    if not node: return []
    return postorder(node.left) + postorder(node.right) + [node.val]

print('In:  ', inorder(root))    # [4,2,5,1,3]
print('Pre: ', preorder(root))   # [1,2,4,5,3]
print('Post:', postorder(root))  # [4,5,2,3,1]
```

```python
# ── Max depth (DFS) ────────────────────────────────────────────────────────
def max_depth(node):
    if not node: return 0
    return 1 + max(max_depth(node.left), max_depth(node.right))

print(max_depth(root))   # 3
```

```python
# ── BFS — level-order traversal ───────────────────────────────────────────
from collections import deque

def level_order(root):
    if not root: return []
    result = []
    q = deque([root])
    while q:
        level = []
        for _ in range(len(q)):       # process one level at a time
            node = q.popleft()
            level.append(node.val)
            if node.left:  q.append(node.left)
            if node.right: q.append(node.right)
        result.append(level)
    return result

print(level_order(root))   # [[1],[2,3],[4,5]]
```

## 12. Graph DFS / BFS

```python
# ── Graph represented as adjacency list ───────────────────────────────────
from collections import defaultdict, deque

graph = defaultdict(list)
edges = [(0,1),(0,2),(1,3),(2,3),(3,4)]
for u, v in edges:
    graph[u].append(v)
    graph[v].append(u)   # undirected
print(dict(graph))
```

```python
# ── Graph DFS (iterative with stack) ──────────────────────────────────────
def dfs(graph, start):
    visited = set()
    stack = [start]
    order = []
    while stack:
        node = stack.pop()
        if node in visited: continue
        visited.add(node)
        order.append(node)
        for nei in graph[node]:
            if nei not in visited:
                stack.append(nei)
    return order

print(dfs(graph, 0))
```

```python
# ── Graph BFS (shortest path) ─────────────────────────────────────────────
def bfs(graph, start, end):
    visited = {start}
    q = deque([(start, 0)])     # (node, distance)
    while q:
        node, dist = q.popleft()
        if node == end:
            return dist
        for nei in graph[node]:
            if nei not in visited:
                visited.add(nei)
                q.append((nei, dist + 1))
    return -1                   # not reachable

print(bfs(graph, 0, 4))  # 2
```

## 13. Dynamic Programming

```python
# ── Fibonacci — 3 ways ─────────────────────────────────────────────────────

# 1. Recursion with memoization (top-down)
from functools import lru_cache

@lru_cache(maxsize=None)
def fib_memo(n):
    if n <= 1: return n
    return fib_memo(n-1) + fib_memo(n-2)

# 2. Tabulation (bottom-up)
def fib_tab(n):
    if n <= 1: return n
    dp = [0] * (n + 1)
    dp[1] = 1
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

# 3. Space-optimized O(1)
def fib_opt(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

for i in range(8):
    print(f'fib({i}) = {fib_memo(i)} | {fib_tab(i)} | {fib_opt(i)}')
```

```python
# ── 0/1 Knapsack ──────────────────────────────────────────────────────────
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            dp[i][w] = dp[i-1][w]                          # skip item i
            if weights[i-1] <= w:
                dp[i][w] = max(dp[i][w],
                               dp[i-1][w - weights[i-1]] + values[i-1])  # take
    return dp[n][capacity]

print(knapsack([2,3,4,5], [3,4,5,6], 5))  # 7
```

```python
# ── Coin Change (unbounded knapsack) ──────────────────────────────────────
def coin_change(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for a in range(1, amount + 1):
        for c in coins:
            if c <= a:
                dp[a] = min(dp[a], dp[a - c] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

print(coin_change([1,5,10,25], 30))  # 2 (25+5)
print(coin_change([2], 3))           # -1
```

## 14. Sorting & Custom Sort

```python
# ── Built-in sort — O(n log n) ─────────────────────────────────────────────
nums = [3,1,4,1,5,9,2,6]
print(sorted(nums))            # new list, ascending
print(sorted(nums, reverse=True))  # descending
nums.sort()                    # in-place

# Sort strings
words = ['banana','apple','cherry']
print(sorted(words))           # alphabetical
print(sorted(words, key=len))  # by length
```

```python
# ── Custom sort with key ───────────────────────────────────────────────────
# Sort intervals by start, then by end descending
intervals = [(1,4),(2,3),(1,5),(3,2)]
intervals.sort(key=lambda x: (x[0], -x[1]))
print(intervals)   # [(1,5),(1,4),(2,3),(3,2)]

# Sort by frequency then value
from collections import Counter
nums = [1,1,2,3,3,3]
freq = Counter(nums)
print(sorted(nums, key=lambda x: (-freq[x], x)))  # [3,3,3,1,1,2]
```

```python
# ── functools.cmp_to_key — for complex comparisons ────────────────────────
from functools import cmp_to_key

# Largest number problem: [3,30,34,5,9] -> "9534330"
def compare(a, b):
    if a + b > b + a: return -1   # a comes first
    if a + b < b + a: return 1
    return 0

nums = ['3','30','34','5','9']
nums.sort(key=cmp_to_key(compare))
print(''.join(nums))   # 9534330
```

## 15. Heaps / Priority Queue

```python
# ── heapq — min-heap by default ───────────────────────────────────────────
import heapq

heap = []
heapq.heappush(heap, 3)
heapq.heappush(heap, 1)
heapq.heappush(heap, 4)
print(heapq.heappop(heap))   # 1 (smallest first)
print(heap)

# Build heap from list in O(n)
nums = [3,1,4,1,5,9]
heapq.heapify(nums)
print(nums)                  # heap order, not sorted
```

```python
# ── Max-heap trick: negate values ─────────────────────────────────────────
import heapq

heap = []
for n in [3,1,4,1,5,9]:
    heapq.heappush(heap, -n)       # negate to simulate max-heap
print(-heapq.heappop(heap))        # 9 (largest)

# Top-k largest elements
nums = [3,1,4,1,5,9,2,6]
print(heapq.nlargest(3, nums))     # [9,6,5]
print(heapq.nsmallest(3, nums))    # [1,1,2]
```

```python
# ── Kth largest element ───────────────────────────────────────────────────
def kth_largest(nums, k):
    # Maintain a min-heap of size k
    heap = nums[:k]
    heapq.heapify(heap)
    for n in nums[k:]:
        if n > heap[0]:            # larger than current min
            heapq.heapreplace(heap, n)
    return heap[0]                 # kth largest = min of heap

print(kth_largest([3,2,1,5,6,4], 2))  # 5
```

## 16. Bit Manipulation

```python
# ── Bit operations ─────────────────────────────────────────────────────────
a, b = 5, 3         # 5 = 101, 3 = 011

print(a & b)        # AND  → 1  (001)
print(a | b)        # OR   → 7  (111)
print(a ^ b)        # XOR  → 6  (110)
print(~a)           # NOT  → -6
print(a << 1)       # left shift  → 10  (multiply by 2)
print(a >> 1)       # right shift → 2   (divide by 2)

print(bin(a))       # '0b101'
print(a.bit_count()) # count set bits → 2  (Python 3.10+)
```

```python
# ── Common bit tricks ──────────────────────────────────────────────────────
n = 12   # 1100

print(n & 1)           # check if odd    → 0 (even)
print(n & (n-1))       # clear lowest set bit → 8 (1000)
print(n & (-n))        # isolate lowest set bit → 4 (0100)

# XOR trick: a ^ a = 0, a ^ 0 = a
# Find the single number in array where every other appears twice
nums = [4,1,2,1,2]
result = 0
for n in nums:
    result ^= n
print(result)          # 4
```

## 17. Useful Python Built-ins

```python
# ── Must-know built-ins ────────────────────────────────────────────────────
from itertools import combinations, permutations, product, accumulate
from functools import reduce
import math

# combinations / permutations
print(list(combinations([1,2,3], 2)))    # [(1,2),(1,3),(2,3)]
print(list(permutations([1,2,3], 2)))    # all ordered pairs

# product (cartesian product / nested loops)
print(list(product([0,1], repeat=2)))    # [(0,0),(0,1),(1,0),(1,1)]

# accumulate (prefix sums)
nums = [1,2,3,4,5]
print(list(accumulate(nums)))            # [1,3,6,10,15]

# reduce
print(reduce(lambda a,b: a*b, [1,2,3,4]))  # 24
```

```python
# ── Math helpers ───────────────────────────────────────────────────────────
import math

print(math.gcd(12, 8))          # 4
print(math.lcm(4, 6))           # 12
print(math.floor(7/2))          # 3
print(math.ceil(7/2))           # 4
print(7 // 2)                   # 3  integer division (same as floor for positives)
print(float('inf'))             # ∞  use for min tracking
print(float('-inf'))            # -∞ use for max tracking
print(math.factorial(5))        # 120
```

```python
# ── zip / map / filter ─────────────────────────────────────────────────────
a = [1,2,3]
b = [4,5,6]

print(list(zip(a, b)))                      # [(1,4),(2,5),(3,6)]
print(list(map(lambda x: x**2, a)))         # [1,4,9]
print(list(filter(lambda x: x%2==0, a)))    # [2]

# zip with unequal lengths — use zip_longest
from itertools import zip_longest
print(list(zip_longest([1,2], [3,4,5], fillvalue=0)))  # [(1,3),(2,4),(0,5)]
```

## 18. Big-O Quick Reference

| Operation | Data Structure | Complexity |
|-----------|---------------|------------|
| Access by index | List | O(1) |
| Append | List | O(1) amortized |
| Insert / Delete (middle) | List | O(n) |
| Lookup | Dict / Set | O(1) avg |
| Insert / Delete | Dict / Set | O(1) avg |
| Push / Pop | Stack (list) | O(1) |
| Enqueue / Dequeue | deque | O(1) |
| Push / Pop | Heap | O(log n) |
| Heapify | Heap | O(n) |
| Sort | Built-in (Timsort) | O(n log n) |
| Binary Search | Sorted list | O(log n) |
| Tree DFS / BFS | Binary tree | O(n) |
| Graph DFS / BFS | Adj list | O(V + E) |

### Space Complexity Rules of Thumb
| Structure | Space |
|-----------|-------|
| Recursion depth | O(depth) stack frames |
| HashMap of n items | O(n) |
| DP table n×m | O(n×m) → often optimizable to O(n) |

## 19. Interview Tips

### During the Interview
1. **Clarify before coding** — ask about input size, duplicates, negative numbers, empty input
2. **State the brute force first** — show you can think, then optimize
3. **Talk through complexity** — always state Time and Space before and after optimizing
4. **Test with examples** — trace through the examples in the problem, then edge cases
5. **Think in patterns** — most LeetCode problems are one of: two pointers, sliding window, BFS/DFS, DP, or binary search

### Pattern Recognition Cheat Sheet
| If the problem asks... | Think... |
|------------------------|----------|
| Sorted array, find pair/triplet | Two pointers |
| Longest/shortest subarray/substring | Sliding window |
| Search in sorted array | Binary search |
| Shortest path in graph | BFS |
| All paths / combinations / permutations | DFS + backtracking |
| Overlapping subproblems | DP |
| Top-K elements | Heap |
| Parentheses, next greater element | Stack |
| Count frequency | HashMap / Counter |
| Duplicate detection | Set |

### Edge Cases to Always Check
- Empty input: `[]`, `""`, `None`
- Single element: `[1]`
- All same elements: `[2,2,2]`
- Negative numbers
- Very large or very small values
